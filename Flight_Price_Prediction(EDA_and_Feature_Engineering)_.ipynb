Flight Price Prediction(EDA and Feature Engineering) 

```python
# importing labraries
import pandas as pd
import numpy as np
import matplotlib as plt
import seaborn as sns
%matplotlib inline
```

```python
train_df=pd.read_excel('/content/Data_Train.xlsx')
train_df.head()
```

```python
test_df=pd.read_excel('/content/Test_set.xlsx')
test_df.head()
```

```python
final_df=train_df.append(test_df)
final_df.head()
```

```python
final_df.tail()
final_df.isnull().sum()
```

```python
final_df.info()
```

```python
#Feature Engineering Process
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
#also u can use this
final_df['Date']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
final_df['Month']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[1])
final_df['Year']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[2])
```

```python
#convert date month and year from object to integer
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df.info()
```

```python
#drop of Date_of_Journey because not in use
final_df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
#Arrival_Time
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
#or
final_df['Arrival_Time']=final_df['Arrival_Time'].apply(lambda x:x.split('/')[0])
```

```python
final_df.head(1)
```

```python
final_df['Arrival_hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_min']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
#convert arrival_hr and arrival_min from object to integer
final_df['Arrival_min']=final_df['Arrival_min'].astype(int)
final_df['Arrival_hour']=final_df['Arrival_hour'].astype(int)
```

```python
final_df.info()
```

```python
#drop the col Arrival_Time 
final_df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Dep_hour']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min']=final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df.head()
```

```python
#convert Dep_hr and Dep_min from object to integer
final_df['Dep_min']=final_df['Dep_min'].astype(int)
final_df['Dep_hour']=final_df['Dep_hour'].astype(int)
```

```python
final_df.info()
```

```python
#drop col dep_time because it is not used
final_df.drop('Dep_Time',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'2 stops':2,'1 stop':1,'3 stops':3,'4 stops':4,'nan':2})
```

```python
final_df.head(5)
```

```python
final_df.drop('Route',axis=1,inplace=True)
```

```python
final_df.head()
```

```python
final_df['Additional_Info'].unique()
```

```python
final_df['Duration_hour']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['Duration_hour']=='5m']
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
#drop duration_hour which is 5m
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hour']=final_df['Duration_hour'].astype(int)
```

```python
final_df.info()
```

```python
#Drop Duration
final_df.drop('Duration',axis=1,inplace=True)
```

```python
final_df.head(1)
```

```python
final_df['Airline'].unique()
```

```python
# Import label encoder
from sklearn import preprocessing
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
 
# Encode labels in column 'species'.
final_df['Airline']= label_encoder.fit_transform(final_df['Airline'])
final_df['Source']= label_encoder.fit_transform(final_df['Source'])
final_df['Destination']= label_encoder.fit_transform(final_df['Destination'])
final_df['Additional_Info']= label_encoder.fit_transform(final_df['Additional_Info'])
 
#df['Airline'].unique()
```
